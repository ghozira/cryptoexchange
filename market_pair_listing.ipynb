```python
url_map_id = 'https://pro-api.coinmarketcap.com/v1/exchange/market-pairs/latest'
parameters_map_id = { 'slug' : 'binance',
                      'limit' : 2000,
                      'category' : 'spot'
                    }
response_map_id = session.get(url_map_id, params = parameters_map_id)
top_xc = pd.json_normalize(json.loads(response_map_id.text)['data']['market_pairs'])
top_xc1 = top_xc[['market_pair_base.exchange_symbol','market_pair','quote.exchange_reported.last_updated','quote.USD.volume_24h']]
top_xc1
```
